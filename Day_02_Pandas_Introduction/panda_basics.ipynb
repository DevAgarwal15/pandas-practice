```python
# ---------------------------------------------
# Load the Zomato Dataset
# ---------------------------------------------

import pandas as pd

df = pd.read_csv("Zomato-data-.csv")

# ---------------------------------------------
# Dataset Information
# ---------------------------------------------

# Display information about the dataset
df.info()

# Generate descriptive statistics
df.describe()

# Count duplicate rows
df.duplicated().sum()

# Plot votes column
df.plot("votes")

# Display 10 random rows
df.sample(10)

# Check the number of rows and columns
df.shape

# Display data types of all columns
df.dtypes

# Display column names
df.columns

# ---------------------------------------------
# Filtering Data
# ---------------------------------------------

# Display restaurants with more than 900 votes
df[df['votes'] > 900]

# ---------------------------------------------
# Data Cleaning
# ---------------------------------------------

# Remove '/5' from ratings
df['rate'] = df['rate'].str.replace('/5', '', regex=False)

# Convert ratings into numeric values
df['rate'] = pd.to_numeric(df['rate'], errors='coerce')

# Display restaurants with rating greater than 4.5
df[df['rate'] > 4.5]

# ---------------------------------------------
# Multiple Condition Filtering
# ---------------------------------------------

# Restaurants offering Dining with table booking
df[(df['listed_in(type)'] == 'Dining') &
   (df['book_table'] == 'Yes')]

# ---------------------------------------------
# Sorting Data
# ---------------------------------------------

# Sort restaurants based on votes
df.sort_values(by='votes', ascending=False)

# ---------------------------------------------
# Rename Columns
# ---------------------------------------------

# Rename 'rate' column to 'rating'
df.rename(columns={'rate': 'rating'}, inplace=True)

# ---------------------------------------------
# Missing Values
# ---------------------------------------------

# Count missing values in the rate column
df['rate'].isnull().sum()
```
