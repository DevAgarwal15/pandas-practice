```python
# ---------------------------------------------------------
# Import required libraries
# ---------------------------------------------------------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# ---------------------------------------------------------
# Load the Zomato dataset
# ---------------------------------------------------------
fd = pd.read_csv("zomato.csv")
```

```python
# ---------------------------------------------------------
# Remove rows containing missing values
# ---------------------------------------------------------
fd.dropna(inplace=True)
# Display the updated shape of the dataset
fd.shape
```

```python
# ---------------------------------------------------------
# Drop unnecessary columns to simplify the dataset
# ---------------------------------------------------------
fd.drop(columns=['name','listed_in(city)','url','address','phone'], inplace=True )
fd[(fd['votes'] > 10000) &( fd['listed_in(type)'] == 'Delivery') & (fd['rate']>'4.5') & (fd['online_order'] == 'Yes')]
```

```python
# ---------------------------------------------------------
# Convert restaurant ratings into numeric values
#
# Replace 'NEW' and '-' with NaN because they are
# not valid numerical ratings.
# ---------------------------------------------------------
def handle_rate(value):
    if(value == 'NEW' or value == '-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = (value[0])
        return float(value)
fd['rate'] = fd['rate'].apply(handle_rate)
# fd['rate'] = fd['rate'].fillna(fd['rate'].mean())
fd.dropna(inplace=True)


# fd['rate'].sort_values(ascending=False).head(10)
```

```python
# ---------------------------------------------------------
# Rename long column names to improve readability
# ---------------------------------------------------------
fd.rename(columns={
    'approx_cost(for two people)':'cost_for_two'

    }, inplace=True)
fd.columns
```

```python
# ---------------------------------------------------------
# Convert commas  into space
# Replace ',' and  with ''
# ---------------------------------------------------------
def convert(value):
    if ';' in value:
        value = value.replace(';','')
        return value
    else:
        return value
    
fd['cuisines'] = fd['cuisines'].apply(convert)
```
